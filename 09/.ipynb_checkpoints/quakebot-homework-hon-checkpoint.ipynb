```python
import pandas as pd
```

```python
earthquakes_df = pd.read_csv("1.0_month.csv")
earthquakes = earthquakes_df.to_dict('records')
```

```python
#Loop through each earthquake, printing sentence descriptions for the ones that are 
#above or equal to 4.0 on the Richter scale.

for item in earthquakes:
    depth = int(item['depth'])
    def depth_to_words(depth):
        if depth <= 70:
            return("shallow")
        if 70 < depth <= 300:
            return("intermediate")
        if depth >= 300:
            return("deep")
    magnitude = int(item['mag'])
    def magnitude_to_words(magnitude):
        if 0 <magnitude <= 4:
            return("like a tickle")
        if 5 <= magnitude <= 7:
            return("a little scary")
        if 8 <= magnitude <= 9:
            return("poop-in-your-pants destructive")
        if magnitude >= 10:
            return("extremely gg bad")
    timestring = item['time']
    yourdate = dateutil.parser.parse(timestring)
    day = (yourdate.strftime("%A"))
    def day_in_words(day):
        return day  
    time = (yourdate.hour) 
    def time_in_words(time):
        if 0 <= time <= 12:
            return("morning")
        if 13 <= time <= 16:
            return("afternoon")
        if 17 <= time <= 19:
            return("evening")
        if time >= 20:
            return("night")
    date = (yourdate.strftime("%B %e"))
    def date_in_words(date):
        return date
    loc = item['place']
    def location_in_words(loc):
        return loc
    def eq_to_sentence(earthquake):
        return "A " + depth_to_words(depth) + ", " + magnitude_to_words(magnitude) + " earthquake was reported " + day_in_words(day) + " " + time_in_words(time) + " on " + date_in_words(date) + " " + location_in_words(loc)

    print(eq_to_sentence(earthquake))
```
